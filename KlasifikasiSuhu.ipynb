```python
def inputSuhu():
    while True:
        masukkan = input('Masukkan suhu (°C) atau "selesai": ')
        
        if masukkan == 'selesai':
            return False

        try:
            suhuC = float(masukkan)
            
            if suhuC < -273.15:
                print('Suhu tidak valid. Dibawah nol absolut (-273.15°C)')
                continue

            return suhuC
        except ValueError:
            print('Input tidak valid. Masukkan angka.')
            continue

def konversiSuhuF(suhuC):
    suhuF = (suhuC * 9/5) + 32
    return suhuF

def konversiSuhuK(suhuC):    
    suhuK = suhuC + 273.15
    return suhuK
    
def kategoriSuhu(suhuC):
    if suhuC < 0:
        return 'Beku'
    elif suhuC <= 15:
        return 'Dingin'
    elif suhuC <= 25:
        return 'Sejuk / Normal'
    elif suhuC <= 35:
        return 'Panas'
    else:
        return 'Sangat Panas'

def cetakSuhu(fh, kv, kd):
    print(f'Fahrenheit : {fh:,.2f}')
    print(f'Kelvin     : {kv:,.2f}')
    print(f'Kondisi    : {kd}')

def rekapSuhu(dt):
    total = 0
    for item in dt:
        total += item

    rataRata = total / len(dt)
    
    print('\n RINGKASAN HARIAN ')
    print(f'Jumlah data    : {len(dt)}')
    print(f'Suhu tertinggi : {max(dt)}')
    print(f'Suhu Terendah  : {min(dt)}')
    print(f'Rata-rata      : {rataRata}')

def main():
    print('===============================')
    print('= SISTEM PENCATAT SUHU HARIAN =')
    print('===============================')
    dataSuhu = []
    while True:
        celcius = inputSuhu()
        if celcius is False:
            break
        fahrenheit = konversiSuhuF(celcius)
        kelvin = konversiSuhuK(celcius)
        kondisi = kategoriSuhu(celcius)
        
        cetakSuhu(fahrenheit, kelvin, kondisi)
        dataSuhu.append(celcius)

    if dataSuhu:
        rekapSuhu(dataSuhu)
    else:
        print('Tidak ada data yang di masukkan')
        
main()
```
